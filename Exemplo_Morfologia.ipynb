```python
#from https://opencv-python-tutroals.readthedocs.io/en/latest/py_tutorials/py_imgproc/py_morphological_ops/py_morphological_ops.html
import cv2 #Importa a biblioteca OpenCv2 que servirá para realizar a leitura das imagens e as operações morfologicas.
import numpy as np #Importa a bibilioetaca Numpy para traduzir as imagens em uma matriz usando prefixo np.
#caso for usar o Google Colab com a OpenCV, usar a lib abaixo.
from google.colab.patches import cv2_imshow #Importa a função para visualização de imagens no google colab.
```

```python
img = cv2.imread('gatos.png',0) #Imagem para realização dos testes
#Opening - Erosão -> Dilatação.
img_opening = cv2.imread('gatos.png',0)#Imagem com ruídos externos ao objetos.
#Closing - Dilatação -> Erosão.
img_closing = cv2.imread('gatos.png',0) #Imagem com ruídos externos ao objetos.
altura, largura = img.shape #Variáveis que constam a altura e a largura da imagem.
kernel = np.ones((5,5),np.uint8) #Define o operador morfologico como sendo uma matriz 5x5 com todos elementos = 1.
print(kernel) #Vizualização do operador morfológico.
```

```python
erosion = cv2.erode(img,kernel,iterations = 2) #versão da imagem erodida com 2 interações.
dilation = cv2.dilate(img,kernel,iterations = 2) #versão da imagem dilatada com 2 interações.
```

```python
gradient = cv2.morphologyEx(img, cv2.MORPH_GRADIENT, kernel) #detecção das bordas das imagens
opening = cv2.morphologyEx(img_opening, cv2.MORPH_OPEN, kernel) #Versão da imagem aberta (processo de erosão seguido pelo de dilatação).
closing = cv2.morphologyEx(img_closing, cv2.MORPH_CLOSE, kernel) #Versão da imagem fechada (Processo de dilatação seguida pelo de erosão).
```

```python
'''
#Caso usa com Python no próprio computador
cv2.imshow('in', img)
cv2.imshow('erosion_out', erosion)
cv2.imshow('dilation_out', dilation)
cv2.imshow('opening_out', opening)
cv2.imshow('closing_out', closing)
cv2.imshow('gradient_out', gradient)
'''
#Caso use o Google Colab
cv2_imshow(img) #Exibe a imagem original.
cv2_imshow(erosion) #Exibe a imagem erodida.
cv2_imshow(dilation) #Exibe a imagem dilatada.
cv2_imshow(opening) #Exibe a imagem aberta (Sem ruídos externos).
cv2_imshow(closing) #Exibe a imagem fechada (Sem ruídos internos).
cv2_imshow(gradient) #Exibe as bordas da imagem.
```
